 Tools: Python, Pandas, NumPy, Matplotlib/Seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('/content/Titanic-Dataset.csv') # importig dataset
df.head() # printing top 4 rows of the data set
```

```python
df.info() # we get complete info about the dataset ex: columns , data types etc
```

```python
print(df.describe()) # stats of the dataset
```

```python
print(df.isnull().sum()) # checking for null values
```

```python
df['Age'].fillna(df['Age'].mean()) # filling null values with mean
```

```python
print(df.isnull().sum()) # checking for null values
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True) # filling null values with mode
```

```python
print(df.isnull().sum()) # checking for null values
```

```python
df['Sex']=df['Sex'].map({'male': 0 , 'female':1}) # converting categorical into numerical
```

```python
df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']]) # feature scaling
```

```python
plt.figure(figsize=(10, 4))
sns.boxplot(data=df[['Age', 'Fare']])
plt.title("Boxplot for Age and Fare")
plt.show()
```

```python
for col in ['Age', 'Fare']:
  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
```

```python
plt.figure(figsize=(10, 4))
sns.boxplot(data=df[['Age', 'Fare']])
plt.title("Boxplot for Age and Fare")
plt.show()
```

```python
df.drop('Sex', axis=1, inplace=True)
```

```python
df.to_excel('titanic_cleaned.xlsx', index=False)
```
